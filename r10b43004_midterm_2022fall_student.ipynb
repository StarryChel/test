### <span style="color:red">Question 1:</span> (5 pts)
Following `hw1_1`, input an integer n>2.
Print a hollow diamond of size n.   
This diamond has double `*` boundary.

<span style="color:green"> <b>Examples:</b></span>
```
>>> hollow_diamon(7)
      *
     ***
    ** **
   **   **
  **     **
 **       **
**         **
 **       **
  **     **
   **   **
    ** **
     ***
      *

>>> hollow_diamon(3)
  *
 ***
** **
 ***
  *
```

````python
# Question 1
def hollow_diamon(n):
    for i in range(1, n+1):
        if i <= 2:
            print(" "*(n - i) + "*"*2*(i-1) + "*")
        else:
            print(" "*(n - i) + "**" + ' '*2*(i-3) + ' ' + "**")
    for i in range(1, n):
       # print(i)
        if i >= n-2:
            print(" "*i + '*'*2*(n-i-1) + "*")
        else:
            print(" "*i + "**" + ' '*(n-2*i+(n-6)) + ' ' + "**")
````

````python
# Q3
def encoder(s):
    ls = list(s)
    a = b = 0
    res=[]
    j = 0
    for i in ls:
        if i == '1':
            a += 1
    for i in ls:
        if i == '0':
            b += 1   
           
    res = str(a) + str(b)     
    return res

print(encoder('11011100'))
````

```python
>>> dora([(1,2),(1,4),(1,5),(5,6),(4,7),(7,8),(2,19),(19,20),(20,23),(23,24)])
5 # 1 --> 2 --> 19 --> 20 --> 23 --> 24. five bridges.
```

<img src="./midterm_picture/Swiper.jpg" width = "400" height = "300" alt="" align=center />


````python
def dora(intervals):
    map_copy = intervals.copy()
    routine = [1]
    
    while 1 > 0:
        box = []
        for i in intervals:
            if routine[-1] in i: #最後一個島
                n = list(i)
                n.remove(routine[-1])
                box.append(min(n))
                map_copy.remove(i)
        intervals = map_copy
        #print(box)
        if box == []: break
        routine.append(min(box))
    return len(routine)
dora([(1,2),(1,4),(1,5),(5,6),(4,7),(7,8),(2,19),(19,20),(20,23),(23,24)])
````
